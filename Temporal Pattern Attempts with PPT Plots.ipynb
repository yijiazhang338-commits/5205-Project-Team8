```python
#import cleaned data
import pandas as pd
df = pd.read_csv("data_clean.csv")
```

<h4>Which audio feature change the quickest</h4>

```python
features = [
    'danceability', 'energy', 'valence', 'tempo',
    'loudness', 'acousticness', 'speechiness',
    'instrumentalness', 'liveness'
]
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

trend_results = {}

for feature in features:
    X = df_yearly['year'].values.reshape(-1, 1)
    y = df_yearly[feature].values
    
    model = LinearRegression().fit(X, y)
    slope = model.coef_[0]
    
    trend_results[feature] = slope
```

```python
import pandas as pd

pd.DataFrame(trend_results.items(), columns=["Audio Feature", "Slope"]) \
  .assign(Abs_Slope=lambda x: x["Slope"].abs()) \
  .sort_values("Abs_Slope", ascending=False) \
  .drop(columns="Abs_Slope") \
  .reset_index(drop=True)
```

```python
#for PPT Plot
import matplotlib.pyplot as plt
import numpy as np

slope_df = (
    pd.DataFrame(trend_results.items(), columns=["Audio Feature", "Slope"])
    .assign(Abs_Slope=lambda x: x["Slope"].abs())
    .sort_values("Abs_Slope", ascending=False)
    .drop(columns="Abs_Slope")
    .reset_index(drop=True)
)

colors = [
    "#1DB954", "#9B5DE5", "#2D8CFF",
    "#FF4D6D", "#FF6B35", "#FFC300",
    "#FFE45E", "#F9DC5C", "#2EC4B6"
]

fig, ax = plt.subplots(figsize=(9, 7), facecolor="#05070D")
ax.set_facecolor("#05070D")

bars = ax.barh(
    slope_df["Audio Feature"],
    slope_df["Slope"],
    color=colors[:len(slope_df)],
    height=0.68
)

ax.axvline(0, color="white", linewidth=1.2, alpha=0.8)

for bar, value, color in zip(bars, slope_df["Slope"], colors):
    y = bar.get_y() + bar.get_height() / 2
    
    if value >= 0:
        ax.text(value + 0.006, y, f"{value:.4f}",
                va="center", ha="left",
                color=color, fontsize=12, fontweight="bold")
    else:
        ax.text(value - 0.006, y, f"{value:.4f}",
                va="center", ha="right",
                color=color, fontsize=12, fontweight="bold")

ax.set_title("Slope (Change per Year)", color="white",
             fontsize=16, fontweight="bold", pad=18)

ax.set_xlabel("Slope", color="white", fontsize=12, fontweight="bold")

ax.tick_params(colors="white", labelsize=12)
ax.grid(False)

for spine in ax.spines.values():
    spine.set_visible(False)

ax.set_xlim(-0.11, 0.20)

ax.invert_yaxis()

plt.tight_layout()
plt.show()
```

<h4>Total Variance Change of Audio Features Over Time

```python
#sum up for the total variance plot
df_var_year = df.groupby('year')[features].var().reset_index()
df_var_year['total_variance'] = df_var_year[features].sum(axis=1)

plt.plot(df_var_year['year'], df_var_year['total_variance'])
plt.title("Overall feature diversity over time")
plt.xlabel("Year")
plt.ylabel("Total variance")
plt.show()
```

```python
#attempts to try different features individually
for feature_name in features:
    df_feature_var_year = (
        df.groupby('year')[feature_name]
          .var()
          .reset_index()
    )

    plt.figure(figsize=(10, 5))
    plt.plot(
        df_feature_var_year['year'],
        df_feature_var_year[feature_name],
        marker='o'
    )

    plt.title(f"Variance of {feature_name} over time")
    plt.xlabel("Year")
    plt.ylabel(f"Variance of {feature_name}")
    plt.grid(True, alpha=0.3)
    plt.show()
```

<h4>PCA Evaluation

```python
#reduce components into 2 using PCA approach
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X = df[features]
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
df_pca = pca.fit_transform(X_scaled)

df['PC1'] = df_pca[:,0]
df['PC2'] = df_pca[:,1]
```

```python
plt.scatter(df['PC1'], df['PC2'], c=df['year'], cmap='viridis')
plt.colorbar(label='Year')
plt.title('PCA Transformation')
plt.show()
```
